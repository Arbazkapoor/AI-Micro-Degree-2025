#### **Step 3: Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Distribution of student grades (G3 - final grade):

```python
plt.figure(figsize=(8,5))
sns.histplot(df['G3'], bins=5, kde=True, color='green')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.title('Distribution of Final Grades')
plt.show()
```

Correlation Heatmap:

```python
plt.figure(figsize=(12,8))
numeric=df.select_dtypes(include='number')
sns.heatmap(numeric.corr(),annot=True,cmap='coolwarm') #corr printing correlation values of all numeric column 
print(numeric.corr())

```

Impact of Study Time on Final Grade:

```python
sns.boxplot(x=df['studytime'],y=df['G3'])  
plt.xlabel("Study Time")
plt.ylabel("Final Grade(G3)")
plt.show()

#weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours , 3 - 5 to 10 hours, or 4 ->10hours)

#10,11,11,11, student avearage studying ,2hrs , 2 to 5 hours,and hours )
```

Parental Education vs. Student Performance:

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Medu'], y=df['G3'])
plt.xlabel("Mother's Education Level")
plt.ylabel('Final Grade')
plt.title("Effect of Mother's Education on Final Grades")
plt.show()
```
